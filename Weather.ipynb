```python
import requests
import json
import datetime
```

```python
print('Current date/time: {}'.format(datetime.datetime.now())) 
```

```python
date_time_str = format(datetime.datetime.now())  
date_time_obj = datetime.datetime.strptime(date_time_str, '%Y-%m-%d %H:%M:%S.%f')

print('Date:', date_time_obj.date())  
print('Time:', date_time_obj.time())  
print('Date-time:', date_time_obj)
```

```python
date = str(date_time_obj.date())
url='https://api.data.gov.sg/v1/environment/2-hour-weather-forecast?date=' + date
r=requests.get(url)
weather = json.loads(r.text)
```

```python
with open('weatherSG.json', 'w') as json_file:  
    json.dump(weather, json_file, indent=1)
```
